### 1. Generate a normally distributed data, with mean=3 and standard deviation=2, for 10,000 observations & Plot the normal distribution overlaid on top of a histogram in matplotlib.

```python
import numpy as np
from scipy.stats import norm
import scipy.stats as st
from matplotlib import pyplot as plt
```

```python
mu, sigma = 3, 2
s = np.random.normal(mu, sigma, 10000)

fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(s, 30, density=True, alpha=0.5, color='b')

xmin, xmax = plt.xlim()
a = np.linspace(xmin, xmax, 100)
p = norm.pdf(a, mu, sigma)
  
plt.plot(a, p, 'r')

plt.show()
```

### 2. Create a sub-sample of this population with mean of 1.5, a sigma of 2.0, and 200 observations & Compare both distributions on a chart..

```python
sample_mean, sample_sigma = 1.5, 2
sample = np.random.normal(sample_mean, sample_sigma, 200)
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
ignored = ax.hist(s, bins=30, alpha=0.2, density=True, color='b')
ignored = ax.hist(sample, 30, alpha=0.2, density=True, color='r')


xmin, xmax = plt.xlim()
a = np.linspace(xmin, xmax, 100)

p = norm.pdf(a, mu, sigma)  
samplep = norm.pdf(a, sample_mean, sample_sigma)  

plt.plot(a, p, 'b')
plt.plot(a, samplep, 'r')

plt.show()
```

### 3. Use from scipy.stats import norm to create 95% confidence intervals. & Create a graph that shows the null and hypotheses, the type 1 and 2 errors

```python
fig, ax = plt.subplots(figsize =(10, 7))
ignored = ax.hist(s, bins=30, alpha=0.2, density=True, color='b')
ignored = ax.hist(sample, 30, alpha=0.2, density=True, color='r')


xmin, xmax = plt.xlim()
a = np.linspace(xmin, xmax, 100)

p = norm.pdf(a, mu, sigma)  
samplep = norm.pdf(a, sample_mean, sample_sigma)  

plt.plot(a, p, 'b')
plt.plot(a, samplep, 'r')

ci = st.t.interval(alpha=0.95, df=len(sample)-1, loc=sample_mean, scale=sample_sigma) 
plt.axvline(ci[0],color='g')
plt.axvline(ci[1],color='g')


plt.fill_between(x=np.arange(-7.5,-2.4,0.01), 
                 y1= st.norm.pdf(np.arange(-7.5,-2.4,0.01), loc= sample_mean, scale=sample_sigma ) ,
                 facecolor='red',
                 alpha=0.7)

plt.fill_between(x=np.arange(-2.44,5.44,0.01), 
                 y1= st.norm.pdf(np.arange(-2.44,5.44,0.01), loc= mu, scale=sigma ) ,
                 facecolor='blue',
                 alpha=0.7)

plt.fill_between(x=np.arange(5.44,10.0,0.01), 
                 y1= st.norm.pdf(np.arange(5.44,10.0,0.01), loc= sample_mean, scale=sample_sigma ) ,
                 facecolor='red',
                 alpha=0.7)


plt.text(x=-2.0, y=0.15, s= "Null Hypothesis", fontsize = 12)
plt.text(x=6.5, y=0.05, s= "Alternative", fontsize = 12)
plt.text(x=-5.0, y=0.01, s= "Type 1 Error", fontsize = 12)
plt.text(x=5.9, y=0.01, s= "Type 1 Error", fontsize = 12)
plt.text(x=2.5, y=0.01, s= "Type 2 Error", fontsize = 12);

plt.show()
```
