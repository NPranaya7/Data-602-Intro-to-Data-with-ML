# SVM

##  Load ‘load_breast_cancer’ dataset from sklearn.datasets.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
cancer_df = load_breast_cancer()
cancer_df
```

## Print the names of the features.

```python
cancer_df.feature_names
```

## Print the names of the labels.


```python
print(cancer_df.keys())
```

```python
df = pd.DataFrame(cancer_df.data)
df
```

```python
df.columns = cancer_df.feature_names
df
```

## Print the shape of the data.

```python
df.shape
```

## Check the top 5 records of the feature set.

```python
df.head(5)
```

## Print the cancer labels for the target set.

```python
target = cancer_df.target
```

```python
df
```

## Split the data (70% training).

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df,target,test_size=0.30,random_state=7)
```

## Create an svm classifier

```python
from sklearn import svm
from sklearn import metrics
```

## Train the model using the training dataset & Predict the response for test dataset.

```python
sm= svm.SVC(kernel='linear')
sm.fit(X_train, y_train)
y_pred = sm.predict(X_test)
print(y_pred)
```

## Evaluate the model with the ‘metrics’ class specifying the ‘precision’ and   ‘recall’ metrics.

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("PRECISION:", metrics.precision_score(y_test, y_pred))
print("Recall:", metrics.recall_score(y_test, y_pred))
```

## What does the recall metric imply?
- Recall metric imply that model is 99% correctely identifies the breast cancer in patients 
- It refers to as True Positive rate of the Model
