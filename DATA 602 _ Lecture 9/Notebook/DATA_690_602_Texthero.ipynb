Visualize the data

```python
import texthero as hero
import pandas as pd
df = pd.read_csv('/content/MIA.csv')
df.head(2)
```

```python
df['text'] = hero.clean(df['text'])
```

```python
df
```

```python
df['tfidf'] = (
    hero.tfidf(df['text'], max_features=100)
)
df[["tfidf", "sentiment"]].head(2)
```

```python
df['pca'] = hero.pca(df['tfidf'])
hero.scatterplot(
    df, 
    col='pca', 
    color='sentiment', 
    title="PCA MIA Engagement"
)
```

```python
df['named_entities'] = (
    hero.named_entities(df['text'])
)
df[['named_entities', 'sentiment']].head(2)
```

```python
NUM_TOP_WORDS = 5
hero.top_words(df['text'])[:NUM_TOP_WORDS]
```

Summary

```python
import texthero as hero
import pandas as pd

df = pd.read_csv('/content/MIA.csv')
df['pca'] = (
    df['text']
    .pipe(hero.clean)
    .pipe(hero.tfidf)
    .pipe(hero.pca)
)

hero.scatterplot(df, col='pca', color='sentiment', title="PCA MIA Engagement")
```
